```python
all_data = pd.DataFrame(columns=['시도별', '년', '월', '재배면적합계', '논_재배면적', '밭_재배면적'])
geo = pd.DataFrame(data.iloc[2:,0].reset_index(drop=True))
geo.columns = ['시도별']
geo
```

```python
all_data
```

```python
data.columns = data.iloc[0]
data = data.drop(data.index[[0,1]]).reset_index(drop=True)
data
```

```python
year = 2002
for i in range(1,54,3):
  new = pd.DataFrame(data.iloc[:,i:i+3])
  new.columns = ['재배면적합계', '논_재배면적', '밭_재배면적']
  temp = pd.concat((geo, new), axis=1)
  temp['년'] = year
  year += 1
  for j in range(1,13):
    temp['월'] = j
    all_data = pd.concat([all_data, temp])
all_data.reset_index(drop=True)
```

```python
data_2002 = pd.read_csv("/content/drive/Shareddrives/2022-1-인공지능/1. 데이터(raw data)/수입량_노지채소_2002.csv", encoding='cp949')
data_2003_2012 = pd.read_csv("/content/drive/Shareddrives/2022-1-인공지능/1. 데이터(raw data)/수입량_노지채소_2003-2012.csv", encoding='cp949')
data_2013_2019 = pd.read_csv("/content/drive/Shareddrives/2022-1-인공지능/1. 데이터(raw data)/수입량_노지채소_2013-2019.csv", encoding='cp949')
display(data_2002, data_2003_2012, data_2013_2019)
```

```python
all_data_income = pd.concat([data_2002.iloc[1:,:],data_2003_2012.iloc[1:,1:], data_2013_2019.iloc[1:,1:]], axis=1) 
all_data_income = all_data_income.transpose()
all_data_income.columns = ['주산물수량','농가수취가격']
all_data_income = all_data_income.drop(all_data_income.index[0])
all_data_income['년'] = all_data_income.index.astype(int)
all_data_income['주산물수량'] = all_data_income['주산물수량'].astype(int)
all_data_income['농가수취가격'] = all_data_income['농가수취가격'].astype(int)
# all_data_income['총수입'] = all_data_income['주산물수량'] * all_data_income['농가수취가격']
all_data_income
```

```python
final_data = pd.DataFrame(columns=['년','월','주산물수량', '농가수취가격'])
for i in range(18):
  for j in range(1,13):
    all_data_income['월'] = j
    final_data = final_data.append(all_data_income.iloc[i])
final_data['총수입량'] = final_data['농가수취가격'] * final_data['주산물수량']
# for col in final_data.columns:
#   final_data[col] = final_data[col].astype(int)
final_data
```

```python
final_data2 = pd.DataFrame()
for i in range(216):
  temp = pd.DataFrame()
  for j in range(17):
    temp = temp.append(final_data.iloc[i])
  temp = temp.reset_index(drop=True)
  final = pd.concat([geo, temp], axis=1, ignore_index=True)
  final_data2 = final_data2.append(final)
final_data2.columns = ['시도별', '년', '월', '주산물수량', '농가수취가격', '총수입량']
for col in final_data2.columns:
  if final_data2[col].dtypes == float:
    final_data2[col] = final_data2[col].astype(int)
final_data2
```

```python
final_data2 = final_data2.reset_index(drop=True)
final_data2
```
